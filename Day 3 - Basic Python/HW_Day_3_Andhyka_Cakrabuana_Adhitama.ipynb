# Basic

1.   Tim data scientist diminta pertolongan oleh CEO ingin meminta beberapa data terkait user. Sehingga tim data scientist diminta membuat data frame dari profil user tersebut. Silahkan buat data frame yang terdiri dari kolom berikut :
  * Nama
  * Tempat lahir
  * Tanggal terakhir berbelanja
  * Rata-rata amount belanja per bulan (Rp)
  * Total transaksi per bulan

2.   Dari data tersebut pilihlah user yang paling loyal berbelanja. DImana user loyal adalah user yang memiliki rata-rata amount belanja per bulan di atas 100 ribu rupiah dan total transaksi per bulan di atas 5.

```python
# Buatlah dataframe yang berisi data 5 user
data_users = {
    'Name' : ['Andi','Budi','Candra','Dodi','Eki'],
    'Birthplace' : ['Jakarta', 'Bogor', 'Depok', 'Tangerang', 'Bekasi'],
    'Birthdate' : ['01-05-1996', '02-06-1995', '03-07-1996', '04-08-1994', '05-09-1997'],
    'LastDateShop' : ['11-01-2020', '22-12-2019', '25-02-2021', '20-11-2022' , '17-08-2021'],
    'AverageShopAmount' : [150000, 120000, 110000, 50000, 30000],
    'MonthlyTotalTransaction' : [6, 4, 7, 8, 3]
}

df_data_users = pd.DataFrame(data_users)
df_data_users
```

```python
# Filter user yang loyal dari data tersebut dimana User loyal = AverageShopAmount > 100000 dan MonthlyTotalTransaction > 5
filter1 = df_data_users['AverageShopAmount'] > 100000
filter2 = df_data_users['MonthlyTotalTransaction'] > 5
filter = filter1 & filter2

df_data_users[filter]
```

**Insight 1** -

**Insight 2** - 
Dengan melihat statistik deskriptif diatas, kita bisa mencari produk terbaik berdasarkan rata-rata Sales dan Profitnya dimana :
*   Rata-rata nilai Sales adalah 221.38
*   Rata-rata nilai Profit adalah 28.21



```python
# Membuat dataframe berdasarkan rata-rata Profit per Category
df_cat_profit = pd.DataFrame(df.groupby(['Category'])['Profit'].mean().sort_values(ascending=False)).reset_index()
df_cat_profit
```

```python
# Melakukan filtering dengan nilai mean dari Profit
filter_cp = df_cat_profit['Profit'] > 28.21
df_cat_profit[filter_cp]
```

**Insight 3** - 
Category produk terbaik dimana secara profit diatas rata-rata adalah Category Technology

```python
# Membuat dataframe berdasarkan rata-rata Profit per Sub-Category
df_scat_profit = pd.DataFrame(df.groupby(['Sub-Category'])['Profit'].mean().sort_values(ascending=False)).reset_index()
df_scat_profit
```

```python
# Melakukan filtering dengan nilai mean dari Profit
filter_scp = df_scat_profit['Profit'] > 28.21
df_scat_profit[filter_scp]
```

**Insight 4** - 
Sub-Category produk terbaik dimana secara profit diatas rata-rata adalah Sub-Category Copiers, Accessories, Appliances, Phones, Chairs.

```python
# Membuat dataframe berdasarkan rata-rata Sales per Category
df_cat_sales = pd.DataFrame(df.groupby(['Category'])['Sales'].mean().sort_values(ascending=False)).reset_index()
df_cat_sales
```

```python
# Melakukan filtering dengan nilai mean dari Sales
filter_cs = df_cat_sales['Sales'] > 221.38
df_cat_sales[filter_cs]
```

**Insight 5** - 
Category produk terbaik dimana secara sales diatas rata-rata adalah Category Technology dan Furniture

```python
# Membuat dataframe berdasarkan rata-rata Sales per Category
df_scat_sales = pd.DataFrame(df.groupby(['Sub-Category'])['Sales'].mean().sort_values(ascending=False)).reset_index()
df_scat_sales
```

```python
# Melakukan filtering dengan nilai mean dari Sales
filter_scs = df_scat_sales['Sales'] > 221.38
df_scat_sales[filter_scs]
```

**Insight 6** - 
Sub-Category produk terbaik dimana secara sales diatas rata-rata adalah Sub-Category Copiers, Machines, Tables, Chairs, Bookcases, Phones, Supplies, Appliances, dan Storage.
